```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titds=pd.read_csv('datasets/TITANICtrain.csv')
titds
```

```python
k=[]
for i in titds["Name"]:
    k.append(i)
k
l=[]
for i in k:
    l.append(i.split(",")[1])
l
j=[]
for i in l:
    j.append(i.split(".")[0])
j
```

```python
titds["titles"]=j
titds["titles"].value_counts()
```

```python
f=[]
for i in titds["titles"].value_counts():
    f.append(i)
len(f)
```

```python
q=[]
for i in titds["Ticket"]:
    q.append(i)
len(q)
l=list(q)
l.sort()
l
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer= TfidfVectorizer() #Term Frequency Inverse Document Frequency, counts number of times a word is repeating in a document. the repetion of the word signifies how important the word actuall is
feature_vectors= vectorizer.fit_transform(l)
```
